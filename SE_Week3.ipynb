```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def quadratic(x,a=-1,b=0,c=10):
  return a*x**2+b*x+c
```

```python
# Define the stages of the Waterfall Model
stages = ["Requirements", "Design", "Implementation", "Testing", "Deployment", "Maintenance"]

# Define a quadratic functiondss
def quadratic(x):
    return -x**2 + 4

# Compute equally spaced x positions and corresponding y positions
x_positions = np.linspace(-2, 2, len(stages))
y_positions = quadratic(x_positions)

# Define a denser set of x values for plotting the curve
x_curve = np.linspace(-2.5, 2.5, 500)
y_curve = quadratic(x_curve)

# Plot the quadratic curve
plt.figure(figsize=(10, 6))
plt.plot(x_curve, y_curve, color="blue", linestyle="--", label="Quadratic Curve")

# Add markers and connect them with vertical lines to create the waterfall effect
for i in range(len(stages)):
    if i > 0:
        # Draw a horizontal segment connecting to the next stage
        plt.plot(
            [x_positions[i-1], x_positions[i]],
            [y_positions[i-1], y_positions[i-1]],
            color="blue", linewidth=2
        )
        # Draw a vertical drop to the next stage
        plt.plot(
            [x_positions[i], x_positions[i]],
            [y_positions[i-1], y_positions[i]],
            color="blue", linewidth=2
        )
    # Add a marker for the stage
    plt.scatter(x_positions[i], y_positions[i], color="red", s=100, zorder=5)
    # Annotate the stage name
    plt.text(
        x_positions[i], y_positions[i] - 0.5, stages[i],
        ha="center", fontsize=10, color="black"
    )

# Customize the plot
plt.title("Waterfall Model Stages", fontsize=14)
plt.xlabel("Stages", fontsize=12)
plt.ylabel("Value (Quadratic)", fontsize=12)
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```
